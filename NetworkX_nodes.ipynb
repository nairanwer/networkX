```python
import networkx as nx
```

```python
#Create an empty graph with no nodes and no edges.
graph = nx.Graph()
#add nodes from any iterable container
graph.add_nodes_from(['a', 'b', 'c', 'd'])
#adding a list of edges
graph.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('a','c')])

nx.draw(graph, with_labels = True)
```

```python
#accessing neighbors
list(graph.neighbors('b'))
```

```python
for neighbor in graph.neighbors('b'):
    print(neighbor)
```

```python
#check for node existence
graph.has_node('a')
```

```python
nx.is_tree(graph)
```

```python
nx.is_connected(graph)
```

```python
#check for number of degrees for each node
graph.degree('a')
```

```python
#directed graphs
DG = nx.DiGraph()
DG.add_edges_from([('1','2'), ('2','3'), ('1','4'), ('3','4')])
nx.draw(DG, with_labels = True)
```

```python
X=nx.Graph()
X.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(X,with_labels=True,node_size=500)
```

```python
dict(X.degree())
```

```python
nx.to_numpy_matrix(X)
```

```python
graph=nx.read_edgelist('graph.txt',nodetype=str,create_using=nx.DiGraph())
nx.draw(graph,with_labels=True,node_size=500)
```
